# 核心算法

```python
#---------------------------------------------------本文算法的函数-----------------------------------------------#
def binary_dual_solve(A1,b1,c1,A2,c2,sparsity=True):
    from scipy.sparse.linalg import eigsh
    # 二分法求对偶问题
    m = A1.shape[0]
    if sparsity:
        A1_temp, b1_temp = np.concatenate((A1.A,b1.A),axis = 1), np.concatenate((b1.A,np.array(c1).reshape(1,1)),axis = 0)
        A = sparse.coo_matrix(np.concatenate((A1_temp,b1_temp.T)))
    else:
        A1_temp, b1_temp = np.concatenate((A1,b1),axis = 1), np.concatenate((b1,np.array(c1).reshape(1,1)),axis = 0)
        A = sparse.coo_matrix(np.concatenate((A1_temp,b1_temp.T)))
        
    mu = eigsh(A,1,which='SA')[0][0]
    lambda_min, lambda_max = 0, (c1-mu)/c2
    epsilon = 1e-6
    df  = 6
    f, g, u = [], [], []
    k, u0, d_lambda = 0, 6, 6

    while abs(d_lambda)>epsilon:
        k = k+1
#         print('k= \n',k)
        lambda_ = (lambda_min+lambda_max)/2
    
        if sparsity:
            A1_temp, b1_temp = np.concatenate(((A1+lambda_*A2).A,b1.A),axis = 1), np.concatenate((b1.A,np.array(c1 - lambda_*c2).reshape(1,1)),axis = 0)
            G = sparse.coo_matrix(np.concatenate((A1_temp,b1_temp.T)))
        else:
            A1_temp, b1_temp = np.concatenate((A1+lambda_*A2,b1),axis = 1), np.concatenate((b1,np.array(c1 - lambda_*c2).reshape(1,1)),axis = 0)
            G = sparse.coo_matrix(np.concatenate((A1_temp,b1_temp.T)))
    #     最小特征值和特征向量
        min_e, min_ev = eigsh(G,1,which='SA')
        
        y = min_ev/min_ev[-1]
        x = y[:-1]
        y = sparse.csr_matrix(y)
        f1, f2 = y.T.dot(A).dot(y), 1+np.linalg.norm(x)**2
        f0 = f1/f2
        f.append(f0)
        
        x = sparse.csr_matrix(x)
        quad_c = x.T.dot(A2).dot(x)
        u0 = (quad_c.A - c2)/(f2)
        u0 = u0[0][0]
        u.append(u0)
        if u0>0:
            lambda_min = lambda_
        else:
            lambda_max = lambda_
        g0, d_lambda = min_e,lambda_max-lambda_min
        g.append(g0)
        df = f0 - g0
    return (g0[0],lambda_)
#-------------------------------------------------------------------------------------------------------------------#
#-----------------------------------------------CVXPY求解问题的函数-------------------------------------------------#
def cvx_dual_solve(A1,b1,c1,A2,c2,sol = 'cvx.SCS',eps=1e-6,sparsity=True):
    # cvx求对偶问题
    solver = {'cvx.SCS':cvx.SCS,'cvx.CVXOPT':cvx.CVXOPT}
    m = A1.shape[0]
    # Construct the problem.
    t = cvx.Variable()
    lambda_ = cvx.Variable()
    objective = cvx.Maximize(t)
    zero = np.zeros(m).reshape(m,1)
#     print(b1.shape)
#     print(zero.shape)
    if sparsity:
        A1_temp, b1_temp = np.concatenate((A1.A,b1.A),axis = 1), np.concatenate((b1.A,np.array(c1).reshape(1,1)),axis = 0)
        B1_temp, b2_temp = np.concatenate((A2.A,zero),axis = 1), np.concatenate((zero,np.array(-c2).reshape(1,1)),axis = 0)
        A = sparse.coo_matrix(np.concatenate((A1_temp,b1_temp.T)))
        B = sparse.coo_matrix(np.concatenate((B1_temp,b2_temp.T)))
#         A = np.insert(np.insert(A1,m,values=b1,axis=0),m,values=np.insert(b1,m,values=c1),axis=1)
#         B = np.insert(np.insert(A2,m,values=zero,axis=0),m,values=np.insert(zero,m,values=-c2),axis=1)
    else:
        A1_temp, b1_temp = np.concatenate((A1,b1),axis = 1), np.concatenate((b1,np.array(c1).reshape(1,1)),axis = 0)
        B1_temp, b2_temp = np.concatenate((A2,zero),axis = 1), np.concatenate((zero,np.array(-c2).reshape(1,1)),axis = 0)
        A = np.concatenate((A1_temp,b1_temp.T))
        B = np.concatenate((B1_temp,b2_temp.T))
        
    constraints = [A+lambda_*B-t*np.eye(m+1)>>0, lambda_ >= 0]
    prob = cvx.Problem(objective, constraints)

    # The optimal objective is returned by prob.solve().
    start = time.clock()
    result = prob.solve(solver=solver[sol],eps=eps)
    elapsed = time.clock() - start
    # The optimal value for x is stored in x.value.
#     print(t.value)
    return (t.value,lambda_.value)
#----------------------------------------------产生随机数据的函数-----------------------------------------------------#
def generate_random_data(m,sparsity = True):
    if sparsity :
        A1, b1, c1 = sparse.random(m,m,density=0.1)*10,sparse.random(m,1,density=0.1)*10,np.random.randint(10)
        A2, c2 = sparse.random(m,m,density=0.1)*10, np.random.randint(1,50)
    else:
        A1, b1, c1 = np.random.randn(m,m)*10,np.random.randn(m)*10,np.random.randint(10)
        A2, c2 = np.random.randn(m,m)*10, np.random.randint(1,50)
    
    A1 = A1.T+A1
#     这样导致A2不是稀疏矩阵
    A2 = A2.T + A2
    A2 = A2 - eigsh(A2,1,which='SA')[0][0]*sparse.eye(m)
    return A1,b1,c1,A2,c2
#---------------------------------------------------------------------------------------------------------------------#
```

# 数值实验
随机产生一组满足条件的矩阵，对比本文方法和一些软件包

```python
# -------------------------------------数值实验，矩阵规模1e2至5e3---------------------------------------------#
N = [10,100,200,500,1000,2000,5000]
# N = [100]
t1,t2,t3 = [],[],[]
dual_binary,dual_scs,dual_cvxopt = [],[],[]
lambda_binary,lambda_scs,lambda_cvxopt = [],[],[]
for n in N:
    A1,b1,c1,A2,c2 = generate_random_data(int(n))
    print("n=",n)
    print("now,binary method is runing")
    t0 = time.clock()
    g,lambda_ = binary_dual_solve(A1,b1,c1,A2,c2)
    t1.append(time.clock()-t0)
    print("now,binary method have finished")
    dual_binary.append(g)
    lambda_binary.append(lambda_)
    
    print("now, SCS is runing")
    t0 = time.clock()
    g,lambda_ = cvx_dual_solve(A1,b1,c1,A2,c2,sol = 'cvx.SCS')
    t2.append(time.clock()-t0)
    print("now, SCS have finished")
    dual_scs.append(float(g))
    lambda_scs.append(float(lambda_))
    
    print("now, CVXOPT is runing")
    t0 = time.clock()
    g,lambda_ = cvx_dual_solve(A1,b1,c1,A2,c2,sol = 'cvx.CVXOPT')
    t3.append(time.clock()-t0)
    print("now, CVXOPT have finished")
    dual_cvxopt.append(float(g))
    lambda_cvxopt.append(float(lambda_))
    
    with open('rq_exp_0523.pickle','wb') as f:
        pickle.dump((t1,t2,t3,dual_binary,dual_scs,dual_cvxopt,lambda_binary,lambda_scs,lambda_cvxopt),f)
```
